```python
import pandas as pd
import glob
```

```python
path = "C:/Users/gaolg/OneDrive/Área de Trabalho/projetos/ips_nordeste/Indicadores"
csv_files = glob.glob(path + "/**/*.csv", recursive=True)
```

```python
path_dict = "C:/Users/gaolg/OneDrive/Área de Trabalho/projetos/ips_nordeste/Dicionario, Documentanção e Codigos/dicionario_regioes.csv"
dicionario = pd.read_csv(path_dict)
dicionario["cod_municipio"] = dicionario["cod_municipio"].astype(str)
dicio = dicionario[["nome_regiao", "sigla_uf", "nome_municipio", "cod_municipio"]]

dicio = dicio.copy()
dicio["cod_municipio6"] = dicio["cod_municipio"].astype(str).str[:6]
dicio_mini = dicio[["cod_municipio6", "cod_municipio"]]
```

```python
keys = []
for file in csv_files:
    key = file.split("\\")[-1].split(".")[0]
    keys.append(key)
```

```python
dataframes = []
for file in csv_files:
    df = pd.read_csv(file)
    dataframes.append(df)
```

```python
for i, df in enumerate(dataframes):
    df.rename(columns={
        "cod_municipio_6": "cod_municipio",
        "cod_municipio6": "cod_municipio"
    }, inplace=True)

    df["cod_municipio"] = df["cod_municipio"].astype(str)

    if len(df["cod_municipio"][0]) == 6:
        df.rename(columns={"cod_municipio": "cod_municipio6"}, inplace=True)
        df = dicio.merge(df, on="cod_municipio6", how="left")
    else:
        df = dicio.merge(df, on="cod_municipio", how="left")

    if "cod_municipio6" in df.columns:
        df = df.drop(columns=["cod_municipio6"])

    # Update the DataFrame in the list with the modified DataFrame
    dataframes[i] = df
```

```python
for i, df in enumerate(dataframes):
    if "cod_municipio6" in df.columns:
        df = df.drop(columns=["cod_municipio6"])

    # Update the DataFrame in the list with the modified DataFrame
    dataframes[i] = df
```

```python
columns_to_check = []
columns_to_remove = ["nome_regiao", "sigla_uf", "nome_municipio", "cod_municipio"]
for df in dataframes:
    column_first = list(set(df.columns.to_list()) - set(columns_to_remove))
    columns_to_check.append(column_first)
```

```python
import itertools
nan_counts = []
for df, column in zip(dataframes, columns_to_check):
     nan_count = df[column[0]].isnull().sum()
     nan_counts.append(nan_count)
    
```

```python
df_nan = pd.DataFrame(list(zip(keys, nan_counts)), columns=["key", "nan_count"])
df_nan.sort_values(by="nan_count", ascending=False)
```
